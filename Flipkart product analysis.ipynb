```python
plt.figure(figsize=(20,5))    # Adjusting the figure size
plt.subplot(1,2,1)            # Plotting two graph together using subplot
flipkart_df['category'].value_counts().plot.pie(autopct='%1.2f%%')     # plotting pie chart category wise 
plt.title('Different Category Available')       # title of first pie chart
plt.subplot(1,2,2)
flipkart_df['out_of_stock'].value_counts().plot.pie(autopct='%1.2f%%')     # Plotting pie chart by out of stock wise
plt.title('Out of Stock')      # title of the second chart
```

```python
# Convert the 'actual_price' column to string first, then remove commas
flipkart_df['actual_price'] = pd.to_numeric(flipkart_df['actual_price'].astype(str).str.replace(',', '', regex=False), errors='coerce')

# Calculate the mean of 'actual_price' (ignoring NaN values)
mean_price = flipkart_df['actual_price'].mean()

# Fill NaN values with the mean price
flipkart_df['actual_price'].fillna(mean_price, inplace=True)
```

```python
flipkart_df['actual_price'] = flipkart_df['actual_price'].astype(int)
```

```python
# Convert 'selling_price' to string first, then remove commas
flipkart_df['selling_price'] = pd.to_numeric(flipkart_df['selling_price'].astype(str).str.replace(',', '', regex=False), errors='coerce')

# Calculate the mean of 'selling_price' (ignoring NaN values)
mean_price = flipkart_df['selling_price'].mean()

# Fill NaN values with the mean price
flipkart_df['selling_price'].fillna(mean_price, inplace=True)

# Convert 'selling_price' to integers
flipkart_df['selling_price'] = flipkart_df['selling_price'].astype(int)
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(x=flipkart_df.index,y='actual_price',data=flipkart_df)
```

```python
plt.figure(figsize=(11,5))
plt.scatter(flipkart_df['actual_price'], flipkart_df['selling_price'], s=50, alpha=0.5)
plt.xlabel('Actual price')
plt.ylabel('selling price')
plt.title('Actual price vs Selling price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# Plot the actual price
plt.plot(flipkart_df['actual_price'], label='Actual Price', marker='o')

# Plot the selling price
plt.plot(flipkart_df['selling_price'], label='Selling Price', marker='o')


plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Actual price vs Selling price')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(flipkart_df['average_rating'], bins=10, edgecolor='k')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Average Ratings')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='sub_category', data=flipkart_df)
plt.xticks(rotation=90)
plt.xlabel('Sub-Category')
plt.ylabel('Count')
plt.title('Count of Products in Each Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x='category', y='average_rating', data=flipkart_df)
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Distribution of Average Ratings by Category')
plt.show()
```
